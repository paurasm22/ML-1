```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import StandardScaler

```

```python
data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv',header=None)
```

```python
data.describe()
```

```python
data.columns=["Pregnancy","Glucose","BloodPressure","SkinThickness","Insulin","Bmi","DiabeticPedigreeFunction","Age","Outcome"]
```

```python
data
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# print("Xtrain is ",x_train,len(x_train))
# print("x_test is ",x_test,len(x_test))
# print("y_train is ",y_train,len(y_train))
# print("y_test is ",y_test,len(y_test))
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)

```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_train_pred = log_reg.predict(X_train)

y_test_pred = log_reg.predict(X_test)
```

```python
# Calculate accuracy on the training set
train_accuracy = accuracy_score(y_train, y_train_pred)

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print the accuracies
print(f'Training Accuracy: {train_accuracy * 100:.2f}%')
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')

# Confusion matrix to understand the prediction distribution
conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:\n", conf_matrix)
```

Now to optimize the accuracy we can train the logistic model on hyperparameters 

```python
# Logistic regression with stronger regularization
log_reg_tuned = LogisticRegression(C=0.1, solver='liblinear')

# Train the model again
log_reg_tuned.fit(X_train, y_train)

# Predict and evaluate again
y_test_pred_tuned = log_reg_tuned.predict(X_test)
tuned_accuracy = accuracy_score(y_test, y_test_pred_tuned)

print(f'Tuned Test Accuracy: {tuned_accuracy * 100:.2f}%')
```
