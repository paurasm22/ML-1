<h1>Prediction of Per Capita Income of Canada using a Linear Regression Model</h1>
<h2>This is Linear Regression with one variable</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv("cppi.csv")
df
```

```python
df.head(4)
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], marker="+", color="red")
```

```python
reg = LinearRegression()
x = df[['year']]
y = df[['per capita income (US$)']]
reg.fit(x, y)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], marker="+", color="red")
plt.xlabel("year")
plt.ylabel("Per Capita Income US$")
plt.plot(x,reg.predict(x),color="blue")
```
