<h1>Canada Per Capita income estimation using polynimial regression</h1>

```python
import numpy as np
import matplotlib.pyplot as plt;
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("cppi2.csv")
df
```

```python
x = df[["year"]].values.reshape(-1,1)
y = df[["pci"]].values

```

```python
plt.scatter(x,y)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x))
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
poly_x = poly.fit_transform(x)
poly_x
```

```python
reg = LinearRegression()
reg.fit(poly_x,y)
```

```python
plt.scatter(x,y , marker="*",color='r')
plt.plot(x,reg.predict(poly_x))
```
