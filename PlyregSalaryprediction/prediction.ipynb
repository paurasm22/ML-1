<h1>Polynomial Regression for prediction of non linear data</h1>

```python
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("sal.csv")
df
```

```python
X = df["Level"].values.reshape(-1,1)
X
```

```python
y = df["Salary"].values
y
```

```python
plot.scatter(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
plot.scatter(X,y)
plot.plot(X,model.predict(X),)
plot.show()
```

```python
model.predict([[3.5]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
# try with 2,3,4 too to understand
poly = PolynomialFeatures(degree=2)
x_poly =poly.fit_transform(X)
x_poly
```

```python
model.fit(x_poly,y)
```

```python
plot.scatter(X,y)
plot.plot(X,model.predict(x_poly),)
plot.xlabel('Levels')
plot.ylabel('Sararies')
plot.show()
```

```python
model.predict([[3.5]])
```

```python
model.predict(poly.fit_transform([[3.5]]))
```
